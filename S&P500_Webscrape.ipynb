```
import bs4 as bs
import requests
import pandas as pd
 
resp = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
soup = bs.BeautifulSoup(resp.text, 'lxml')
table = soup.find('table', {'class': 'wikitable sortable'})
if len(table.contents) == 0:
  print("No table found")
else:
  print("Table extracted correctly!")
  print(type(table))
```

```
tickers = []
industries = []
for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text
        #fourth element is the sector
        industry = row.findAll('td')[4].text
        
        tickers.append(ticker)
        industries.append(industry)
```

```
print(tickers)
```

```
tickers = list(map(lambda s: s.strip(), tickers))
industries = list(map(lambda s: s.strip(), industries))

print(tickers)
```

```
tickerdf = pd.DataFrame(tickers,columns=['ticker'])
sectordf = pd.DataFrame(industries,columns=['industry'])
```

```
tickerandsector =pd.merge(tickerdf, sectordf, on = tickerdf.index, how = "outer")
tickerandsector = tickerandsector.drop('key_0', axis=1)
```

```
tickerandsector
```

```
ind = pd.DataFrame(tickerandsector['industry'].unique(), columns=['Industry Sector'])
print(ind)
```

```
filtersector = tickerandsector.loc[tickerandsector['industry'] == 'Industry Sector']

listoftickers =  filtersector['ticker'].tolist()
print(listoftickers)
```

```
#mirror tickerdf Datafram to Stock Symbol
stock_symbol = tickerdf
tickerdf
```

```
#test run to show only first 50 stocks
price_close = []
value = []

for j in range(len(stock_symbol[1:50])):
    link = 'https://finance.yahoo.com/quote/'+stock_symbol["ticker"].iloc[j]+'?p='+stock_symbol["ticker"].iloc[j]
    #webbrowser.open(link)
    page = requests.get(link)
    html_soup = bs.BeautifulSoup(page.content, 'html.parser')
    price_close.append(html_soup.find('span', attrs = {'class' : 'Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(ib)'}).text)
    page = requests.get(link)
    html_soup = bs.BeautifulSoup(page.content, 'html.parser')
    #price_close.append(html_soup.find('span', attrs = {'class' : 'Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(ib)'}).text)
    x = html_soup.find('div', attrs = {'class' : 'Fw(b) Fl(end)--m Fz(s) C($primaryColor'}).get_text()
    if 'Overvalued' in x:
        value.append('Overvalued')
    elif 'Undervalued' in x:
        value.append('Undervalued')
    elif 'Nuetral' in x:
        value.append('Nuetral')
    elif 'Near Fair Value' in x:
        value.append('Near Fair Value')
    else:
        value.append('N/A')



price_close = pd.DataFrame(price_close, columns=['Closing Price'])
value = pd.DataFrame(value, columns=['Value'])

print(value)
print("\n")
print(price_close)
print("\n")
print(tickerdf.iloc[range(len(stock_symbol[1:50]))])
```

```
df = pd.merge(price_close, tickerdf, left_index=True, right_index=True)
df
```

```
df_other =  pd.merge(df , value, left_index=True, right_index=True, sort = True)
df_other
```

```
df_final = pd.merge(ind , df_other, left_index=True, right_index=True, sort = True)
df_final.sort_values(by=['Industry Sector'], ascending=True)
```

```
import datetime
x = datetime.datetime.now()
name = x.strftime("%a_%d_%y_") + "S&P500"

#saves onto excel file
df_final.to_excel("Portfolio.xlsx", sheet_name=name, index=False)
print("Excel file created Succesfully!")
```
